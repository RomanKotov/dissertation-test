```python
# Implementation of matplotlib function
import matplotlib.pyplot as plt
import numpy as np

feature_x = np.linspace(-3.0, 3.0, 170)
feature_y = np.linspace(-3.0, 3.0, 170)

alfa = 0.5

def MyAnd(x,y):
    return (1/(1+alfa))*(x+y-(x*x+y*y-2*alfa*x*y)**0.5)
def MyOr(x,y):
    return (1/(1+alfa))*(x+y+(x*x+y*y-2*alfa*x*y)**0.5)
# Creating 2-D grid of features
[X, Y] = np.meshgrid(feature_x, feature_y)

fig, ax = plt.subplots(1, 1)

def f1(x,y):
   return 4-(1-x)**2-y**2
def f5(x,y):
   return 4-x**2-y**2
    
def f2(x,y):
   return 4-(1+x)**2-y**2

def f3(x,y):
   return y
    
def f4(x,y):
   return x    
    
def Res(x,y):
   v1 = f1(x,y)
   v2 = f2(x,y)
   v3 = f3(x,y) 
   v4 = f4(x,y)  
   v5 = f5(x,y)   
   #return MyAnd(v4,MyAnd(v3,v5))
   #return v5 
   return MyOr(v1,v2)


cs = ax.contourf(X, Y, Res(X,Y), levels=10)
cb=plt.colorbar(cs)
ax.set_title('Filled Contour Plot')
ax.set_xlabel('feature_x')
ax.set_ylabel('feature_y')

plt.show()
```
